# Département d'architecture : construction d'une pyramide ---Validation---

<u><b>Répétitions conditionnées<b><u>

lien : http://www.france-ioi.org/algo/task.php?idChapter=649&idTask=2052

```python
#code : 
nbPierresMax = int(input())
hauteur = 0
nbPierresNecessaire = 0
while (nbPierresNecessaire + (hauteur+1)*(hauteur+1)) <= nbPierresMax:
   hauteur = hauteur+1
   nbPierresNecessaire += (hauteur*hauteur)
print(hauteur)
print(nbPierresNecessaire)
```
